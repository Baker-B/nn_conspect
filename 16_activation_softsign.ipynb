### 3.3 What is Softsign activation function and its derivative?

In this post, we will talk about the Softsign activation function and its derivative. The shape of the Softsign function is very similar to the Sigmoid function but the output range is (-1, 1) unlike (0, 1) which is for the Sigmoid.

This is the definition of the Softsign function.

```python
%%latex
\begin{gather*}
    y = softsign(x) = f(x) = \frac{x}{| x |+ 1} \\
        \\
        \text{And it is very easy to find the derivative of the Softsign function.}\\
        \\
        \frac{dy}{dx} = f'(x) = \frac{d(\frac{x}{| x |+ 1})}{dx} \Rightarrow \\
        
        \\
        \Rightarrow f'(x) = \frac{1 (| x | + 1) - x(\frac{x}{|x|})}{(| x | + 1)^{2}} \Rightarrow \\
        \\
        \Rightarrow f'(x) = \frac{| x | + 1 - \frac{x^{2}}{|x|}}{(| x | + 1)^{2}} \Rightarrow \\
        \\
        \Rightarrow f'(x) = \frac{| x | \frac{|x|}{|x|} + 1 - \frac{x^{2}}{|x|}}{(| x | + 1)^{2}} \, \color{red} \text{How???}\Rightarrow \\
        \\
        \Rightarrow f'(x) = \frac{\frac{x^{2}}{|x|} + 1 - \frac{x^{2}}{|x|}}{(| x | + 1)^{2}} \Rightarrow \\
        \\   
        \Rightarrow f'(x) = \frac{1}{(| x | + 1)^{2}}
        \\
        \color{green} \text{This may help if you are wondering how to take the derivative of absolute x.}
        \\
        y = f(x) = |x| = \sqrt{x^{2}} = (x^{2})^{\frac{1}{2}} \\
        \\
        \frac{dy}{dx} = f'(x) = \frac{d((x^{2})^{\frac{1}{2}})}{dx} \Rightarrow\\
        \\
        \Rightarrow f'(x) = \frac{1}{2}(x^{2})^{\frac{-1}{2}}(2x) \Rightarrow\\
        \\
        \Rightarrow f'(x) = x(x^{2})^{\frac{-1}{2}} Rightarrow\\
        \\
        \Rightarrow f'(x) = \frac{x}{(x^{2})^{\frac{1}{2}}} Rightarrow\\
        \\
        \Rightarrow f'(x) = \frac{x}{|x|}
            
        
\end{gather*}
```

We can easily implement the Softsign function in Python.

```python
import numpy as np                             # importing NumPy
np.random.seed(42)

def softsign(x):                               # Softsign
    return x / (abs(x) + 1)

def softsign_dash(x):                          # Softsign derivative
    return 1 / (abs(x) + 1)**2
```

Let us have a look at an example.

```python
x = np.array([[0.2], [0.5], [1.2], [-2.3], [0]])
print(x)
print(softsign(x))
softsign_dash(x)
```
